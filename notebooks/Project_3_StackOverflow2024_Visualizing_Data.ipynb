- Display columns and their values

```python
def check_columns(df:pd.DataFrame) -> None:
    # Identify feature types
    numeric_features = df.select_dtypes(include=['number']).columns.tolist()
    boolean_features = df.select_dtypes(include=['bool']).columns.tolist()
    categorical_features = df.select_dtypes(include=['object']).columns.tolist()

    print('>> Info :')
    df.info()
    print('>> Columns : ')
    print(f'> Numeric Features     is {numeric_features}')
    print(f'> Boolean Features     is {boolean_features}')
    print(f'> Categorical Features is {categorical_features}')
    print('>> Values : ')
    for feature in df.columns:
        print(f'> {feature} : {df[feature].unique().tolist()}')
```

## Step 4 : Visualization

- Define a function to plot line, bar and pie charts

```python
def plot_charts(df:pd.DataFrame, column:str, top_n:int=20, class_list:list=[]):
    # Check if the specified column exists in the DataFrame
    if column not in df.columns:
        raise ValueError(f"Column '{column}' does not exist in the DataFrame.")
    
    # Group by the class column and sum the value column
    if len(class_list) == 0:
        class_col = df[column].value_counts().keys().tolist()[0:top_n]
        value_col = df[column].value_counts().values.tolist()[0:top_n]
    else:
        class_col = class_list
        value_col = df[df[column].isin(class_list)].tolist()
    
    # Create a DataFrame for plotting
    data_to_plot = pd.DataFrame({column : class_col, 'Count' : value_col}).reset_index(drop=True)
    
    if column not in df.select_dtypes('number').columns.tolist():
        # Line chart
        fig_line = px.line(data_frame=data_to_plot, x=column, y='Count', title='Line Chart')
        fig_line.show()

    # Bar chart
    fig_bar = px.bar(data_frame=data_to_plot, x=column, y='Count', title='Bar Chart')
    fig_bar.show()

    # Pie chart
    fig_pie = px.pie(data_frame=data_to_plot, names=column, values='Count', title='Pie Chart')
    fig_pie.show()
```

- Visualize Features

```python
ntop = 10 
for index, col in enumerate(df.columns.tolist()):
    print(f"> {index + 1} : {col} ...")
    plot_charts(df, col, ntop)
```
